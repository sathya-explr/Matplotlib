```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        startangle=30, autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'black'})

# explode makes a particular segment/slice pop out slightly 
# shadow parameter gives a 3D effect
# wedgeprops gives border
# start angle rotates the piechart
# autopct gives the percentage values

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
