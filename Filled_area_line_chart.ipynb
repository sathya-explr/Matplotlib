```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head(5)
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
print(type(ages))
```

```python
# a simple line chart

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

plt.legend()

# plt.legend is for showing the label data for ech line

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

#### Filling the region under a single line. (The y2  parameter is not mentioned here) Therefore the grph will fill the entire area under it



```python
plt.fill_between(ages, py_salaries)

plt.plot(ages, py_salaries, label='Python')

plt.legend()
```

#### Alpha parameter

```python
# To make the graph lighter, so that we could see thru'the graph, we use parameter alpha


plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, alpha = 0.25)


plt.legend()
```

#### Filling up and down, setting the y limit with y2 parameter.

```python
overall_median = 57287

plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, overall_median, alpha = 0.25 )

plt.legend()
```

#### Making a conditional while plotting the graph, using where clause

```python
plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, overall_median, where= (py_salaries >  overall_median), alpha = 0.25 )


plt.legend()

plt.tight_layout()
```

#### Use interpolate, so that all the regions are filled correctly

```python
plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, overall_median, where= (py_salaries >  overall_median),interpolate = True, alpha = 0.25 )


plt.legend()

plt.tight_layout()
```

```python
# For getting different colours for region above and region below median, we need to use the fill_between method twice.

plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, overall_median, where= (py_salaries >  overall_median),interpolate = True, alpha = 0.25 )

plt.fill_between(ages, py_salaries, overall_median, where= (py_salaries <=  overall_median),interpolate = True, alpha = 0.25 )

plt.tight_layout()
```

#### label in fill_between method

```python
# to compare the two salaries

plt.plot(ages, py_salaries, label='Python')

plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.fill_between(ages, py_salaries, dev_salaries, where= (py_salaries >  dev_salaries),interpolate = True, alpha = 0.25,label='Greater than avg dev salary'  )

plt.fill_between(ages, py_salaries, dev_salaries, where= (py_salaries <=  dev_salaries),interpolate = True, alpha = 0.25, label = 'Below Dev')


plt.legend()

plt.tight_layout()
```
