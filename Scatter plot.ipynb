```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

#### Simple Scatter Plot

```python
plt.scatter(x,y)
```

```python
# In the scatter fn you introduce a s parameter to change the size of the dots, c to change color, marker to change the dot
plt.scatter(x,y, s=100, c = 'green', marker = 'x')
```

[Matplotlib styles](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html),


[Matplotlin colormaps](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)

 Alpha( to make the pts lighter), edge color (to give a boundry to the pts) and linewidth decides the width of the boundry

```python
plt.scatter(x,y, s=100, c = 'green', edgecolor = 'black', linewidth = 1, alpha = 0.55)
```

#### To differentiate the points based on intensity :
We use parameter colors and give a list of number here, 

```python
clrs = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x,y, s=100, c = clrs, edgecolor = 'black', linewidth = 1, alpha = 0.55)

```

```python
#cbar = plt.colorbar()
#cbar.set_label('satisfactory')

# we use cmap parameter to change dot colors from shades of grey


plt.scatter(x,y, s=100, c = clrs, cmap = 'Reds', edgecolor = 'black', linewidth = 1, alpha = 0.55)
```

```python
####  to sepcify what they (intensities) represent , we give a label in the colorbar method (legend)

plt.scatter(x,y, s=100, c = clrs, cmap = 'Reds', edgecolor = 'black', linewidth = 1, alpha = 0.55)

cbar = plt.colorbar()
cbar.set_label('satisfactory')

# it is like color coding, 
#there is no need to use plt.legend here
```

#### To vary the size of dotd, we can a list to the s parameter of scatter fn

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]


plt.scatter(x,y, s=sizes, c = clrs, cmap = 'Reds', edgecolor = 'black', linewidth = 1, alpha = 0.55)

cbar = plt.colorbar()
cbar.set_label('satisfactory')
```

```python
data = pd.read_csv('Scatterdata.csv')
```

```python
data.head(5)
```

```python
view_count = data['view_count']
likes = data['likes']
ltd_ratio = data['ratio']
```

```python
# A scatter between view_counts and likes

plt.scatter(view_count,likes,edgecolor = 'black', linewidth = 1, alpha = 0.55 )
```

#### As the above graph is concentrated, we use a log scale to check out the distribution

```python
# log scale in scatter plots

plt.scatter(view_count,likes,edgecolor = 'black', linewidth = 1, alpha = 0.55 )

plt.xscale('log')
plt.yscale('log')
```

```python
# To give a colored intensity, we use the c parameter, and here the c is based on the ltd_ratio, to give a laber we use cbar 
# and for choosing a color pattern we use the cmap parameter

plt.scatter(view_count, likes, c=ltd_ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')


plt.xscale('log')
plt.yscale('log')

#Adding title, x and y labels

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

```
